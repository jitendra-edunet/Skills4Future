```python
# Scaling 

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)
```

```python
# Parametreler
length = 12   
batch_size = 1   
stride = 1  
# Veri ve hedef oluşturma
data_2 = []
targets = []

for i in range(0, len(train_scaled) - length + 1, stride):
    data_2.append(train_scaled[i:i+length])
    targets.append(train_scaled[i+length-1])

data_2 = np.array(data_2)
targets = np.array(targets)
```

```python
print(data_2)
```

```python
print(targets)
```

```python
train_scaled.shape
```

```python
n_features = train_scaled.shape[1]
n_features
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN, Bidirectional, Dropout
```

```python
from keras.layers import Dropout

model = Sequential()
model.add(SimpleRNN(units=128, activation="tanh", input_shape=(length, n_features)))
model.add(Dropout(0.2))  # Dropout layer added, dropout rate set to 0.2
model.add(Dense(64, activation="tanh"))  # "relu" was used as activation function
model.add(Dropout(0.2))
model.add(Dense(32, activation="tanh"))  # "tanh" was used as activation function
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer="adam", loss="mse")
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model.fit(data_2, targets, epochs = 20, batch_size=batch_size)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot();
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
train_scaled.shape
```

```python
n_features = train_scaled.shape[1]
n_features
```

```python
model = Sequential()
model.add(LSTM(64, activation = 'tanh', return_sequences=True, input_shape = (length, n_features)))
model.add(Dropout(0.2))  # Dropout layer added, dropout rate set to 0.2
model.add(LSTM(32, activation = 'tanh', return_sequences=True))
model.add(Dropout(0.2))  # Dropout layer added, dropout rate set to 0.2
model.add(LSTM(16, activation = 'tanh'))
model.add(Dropout(0.2))  # Dropout layer added, dropout rate set to 0.2
model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', patience = 5, restore_best_weights = True)
```

```python
model.summary()
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
validation_generator = TimeseriesGenerator(test_scaled, test_scaled, length = length, batch_size = batch_size)
```

```python
generator = TimeseriesGenerator(train_scaled, train_scaled, length = length, batch_size = batch_size)   # shuffle= false
```

```python
model.fit(generator,
            epochs = 20,
            validation_data = validation_generator,
            callbacks = [early_stop]);
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot();
```
